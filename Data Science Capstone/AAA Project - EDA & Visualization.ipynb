```python
# filter to numerical columns
numericalColumns = upgrade_df.dtypes[upgrade_df.dtypes!=object]

num_columns=list(numericalColumns.index)

num_upgrade_df=upgrade_df[num_columns]
```

```python
# heatmap for the first 20 numerical variables

correlation=round(num_upgrade_df[num_upgrade_df.columns[2:11]].corr(),2)

# change the size of the plot

plt.gcf().set_size_inches(10, 10)

#sns.heatmap(correlation)

sns.heatmap(correlation, cmap="YlGnBu", annot=True)
```
